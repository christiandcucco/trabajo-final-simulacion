## Ventanas de período 3

```python
def mann_orbit(x,a,lam):
    return ((1-a)*x+(a*lam*x*(1-x)))

def mann_orbit_3_period(x, a, lam):
    return mann_orbit(mann_orbit(mann_orbit(x,a,lam),a,lam),a,lam)
```

Grafico del mapa con ventana de período 3 con lambda = 4.15 y a = 0.9

```python
x_results_mann_orbit = []
y_results_mann_orbit = []
x_axis = []
y_axis = []

lam = 4.15
a = 0.9
x = 0

while x <= 1 :
    y_results_mann_orbit.append(mann_orbit_3_period(x,a,lam))
    x_results_mann_orbit.append(x)
    x_axis.append(x)
    y_axis.append(x)
    x += 0.01
plt.figure(figsize=(20,10))
plt.plot(x_results_mann_orbit,y_results_mann_orbit)
plt.plot(x_axis,y_axis)

plt.show()
```

## Exponente Lyapunov

```python
def mann_orbit_derivate(a,lam,x):
    result = (1-a+(a*lam)-(2*a*lam*x))
    if result == 0:
        result += 0.0000001
    return result

def process_x_results(x_results, total, a, lam):
    acumulated = 0
    for i in range(total):
        acumulated += (math.log(abs(mann_orbit_derivate(a,lam,x_results[i]))))
    return acumulated
    
```

###### Exponentes de Lyapunov para la órbita x0 = 0.3, lambda entre 0 y 3.2199 y a = 0.9

```python
x_results1 = []
x_results2 = []
x_results3 = []

lyapunov_results1 = []
lyapunov_results2 = []
lyapunov_results3 = []

max_iterations = 100

lam1 = 3
lam2 = 3.1
lam3 = 2.5

x1 = 0.3
x2 = 0.3
x3 = 0.3

a = 0.9

for i in range(max_iterations):
    divisor = i
    if divisor == 0:
        divisor = 1

    x_results1.append(x1)
    x_results2.append(x2)
    x_results3.append(x3)

    lyapunov_results1.append((process_x_results(x_results1, i, a, lam1))/divisor)
    lyapunov_results2.append((process_x_results(x_results2, i, a, lam2))/divisor)
    lyapunov_results3.append((process_x_results(x_results3, i, a, lam3))/divisor)

    x1 = mann_orbit(x1,a,lam1)
    x2 = mann_orbit(x2,a,lam2)
    x3 = mann_orbit(x3,a,lam3)

plt.figure(figsize=(20,10))
plt.plot(lyapunov_results1,'g')
plt.plot(lyapunov_results2,'r')
plt.plot(lyapunov_results3,'c')
plt.show()
```

###### Variación del exponente de Lyapunov respecto del valor lambda

```python
def process_lambda_variation_lyapunov_exponent(x, a, lam):
    max_iterations = 1000
    x_results = []
    lyapunov_exponent = 0

    for i in range(max_iterations):
        divisor = i
        if divisor == 0:
            divisor = 1   
        x_results.append(x)
        lyapunov_exponent = process_x_results(x_results, i, a, lam)/divisor
        x = mann_orbit(x,a,lam)
    return lyapunov_exponent
```

###### Exponente de Lyapunov  para lambda entre 1 y 21 con a = 0.1

```python
max_lambda = 21
current_lambda = 1
lyapunov_results = []
lambdas = []
a = 0.1
x = 0.3

while current_lambda <= max_lambda:
    lyapunov_results.append(process_lambda_variation_lyapunov_exponent(x,a,current_lambda))
    lambdas.append(current_lambda)
    current_lambda +=0.1

plt.figure(figsize=(20,10))
plt.plot(lambdas,lyapunov_results)
plt.show()
```
